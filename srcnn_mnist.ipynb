```
from keras import backend as K
from keras.models import Sequential
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.core import Activation
from keras.layers.core import Flatten
from keras.layers.core import Dense
from keras.datasets import mnist
from keras.utils import np_utils
from keras.optimizers import Adam
from keras.preprocessing.image import array_to_img, img_to_array,  load_img
import numpy as np
import cv2
```

```
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```
y_train[0]
```

```
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255 
x_train = x_train[:, :, :, np.newaxis]
x_test = x_test[:, :, :, np.newaxis]
```

```
x_train.shape
```

```
x_test.shape
```

```
input=[]
x_train5 = []

for i in range(60000):
  if y_train[i]==5:
    sample=x_train[i].copy()
    x_train5.append(x_train[i])
    sample=cv2.resize(sample, dsize=(14,14))
    sample=cv2.resize(sample, dsize=(28,28), interpolation=cv2.INTER_NEAREST)
    input.append(sample)
```

```
test=[]
x_test5 = []

for i in range(10000):
  if y_test[i]==5:
    sample_t=x_test[i].copy()
    x_test5.append(x_test[i])
    sample_t=cv2.resize(sample_t, dsize=(14,14))
    sample_t=cv2.resize(sample_t, dsize=(28,28), interpolation=cv2.INTER_NEAREST)
    test.append(sample_t)
```

```
input = np.array(input)
x_train5 = np.array(x_train5)
```

```
x_train5.shape
```

```
input.shape
```

```
input = input.reshape(5421,28,28,1)
```

```
modelx2 = Sequential()
modelx2.add(Conv2D(16, (3, 3), activation='relu', padding='same', input_shape=(28,28,1)))
modelx2.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
modelx2.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
modelx2.add(Conv2D(1, (3, 3), activation='relu', padding='same'))

modelx2.summary()
```

```
modelx2.compile(optimizer='rmsprop', loss='mse', metrics=['accuracy'])
```

```
modelx2.fit(input, x_train5, epochs=20, batch_size=64)
```
